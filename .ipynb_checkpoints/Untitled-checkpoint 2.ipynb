```python
def addInfoEstato(state):
    global estadoscovid
    EvolucaoEstado = df_merge_difkey[df_merge_difkey.state == state].groupby('date').sum()['confirmed'].reset_index()
    EvolucaoEstado = EvolucaoEstado.drop('date',axis=1)
    EvolucaoEstado.columns = [state]
    estadoscovid = pd.concat([estadoscovid, EvolucaoEstado], ignore_index=False,axis=1)
```

```python
def addInfoRegiao(regiao):
    global Regiaocovid
    EvolucaoRegiao = df_merge_difkey[df_merge_difkey['Região'] == regiao].groupby('date').sum()['confirmed'].reset_index()
    EvolucaoRegiao = EvolucaoRegiao.drop('date',axis=1)
    EvolucaoRegiao.columns = [regiao]
    Regiaocovid = pd.concat([Regiaocovid, EvolucaoRegiao], ignore_index=False,axis=1)
```

```python
estadoscovid = pd.DataFrame()
Regiaocovid = pd.DataFrame()


for est in estados:
    addInfoEstato(est)
    
for reg in Regiao:
    addInfoRegiao(reg)
    
```

```python
##como a correção entre confimed e deaths é alta, vamos montar uma predição usando regração linear
##neste caso vamos montar um novo dado onde teremos as duas variaveis uma será o numero de casos x mortes
## desta forma teremos a variavel X e ˆx.
dfPred = pd.DataFrame()

EvolucaoEstado = df_merge_difkey[df_merge_difkey.state == 'SP'].groupby('date').sum()['confirmed'].reset_index()
EvolucaoEstado = EvolucaoEstado.drop('date',axis=1)
EvolucaoEstado.columns = ['confirmed']
dfPred = pd.concat([dfPred, EvolucaoEstado], ignore_index=False,axis=1)

EvolucaoEstado = df_merge_difkey[df_merge_difkey.state == 'SP'].groupby('date').sum()['deaths'].reset_index()
EvolucaoEstado = EvolucaoEstado.drop('date',axis=1)
EvolucaoEstado.columns = ['deaths']
dfPred = pd.concat([dfPred, EvolucaoEstado], ignore_index=False,axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

import statsmodels.api as sm


def calcPolyTest(X_sm,degree, lm_model):
    poly = PolynomialFeatures(degree=degree)
    poly_X = poly.fit_transform(X_sm)
    y_hat = lm_model.predict(poly_X)
    return y_hat

def calcPoly(y, X_sm,degree, lm_model):
    poly = PolynomialFeatures(degree=degree)
    poly_X = poly.fit_transform(X_sm)
    lm_model.fit(poly_X, y)
    y_hat = lm_model.predict(poly_X)
    return y_hat
```

```python
X = dfPrednew.confirmed.head(50).values
y = dfPrednew.deaths.head(50).values

X_teste = dfPrednew.confirmed.head(70).values
y_controle = dfPrednew.deaths.head(70).values
```

```python
from sklearn.metrics import r2_score

#X_sm = sm.add_constant(X)
X_sm = X.reshape(-1,1)
X_sm_test = X_teste.reshape(-1,1)

print(X_sm.shape)
print(X_sm_test.shape)

lm_model = LinearRegression()
y_hat_1 = calcPoly(y,X_sm,2,lm_model);
print('Finalizado 1')

lm_model2 = LinearRegression()
y_hat_2 = calcPoly(y,X_sm,3,lm_model2);
print('Finalizado 2')

lm_model3 = LinearRegression()
y_hat_3 = calcPoly(y,X_sm,4,lm_model3);
print('Finalizado 3')

lm_model4 = LinearRegression()
y_hat_4 = calcPoly(y,X_sm,5,lm_model4);
print('Finalizado 4')

lm_model5 = LinearRegression()
y_hat_5 = calcPoly(y,X_sm,8,lm_model5);
print('Finalizado 5')


y_hat_test = calcPolyTest(X_sm_test[:50],4,lm_model3)
print(X_sm.shape)
print(y_hat_test.shape)

r2 = r2_score(y_controle[:50], y_hat_test) #Acuracia
print(r2)

#results = sm.OLS(y, X_sm).fit()
#Yols_hat = results.predict(X_sm)
```

## Erro absoluto médio (mae) é a medida da diferença entre duas variáveis contínuas
## Seja Y os valores reais e ˆy os valores previstos de um determinado evento
## MAE É A DISTANVIA VERTICAL MEDIA ENTRE DOS PONTOS
## MAE UTILIZA A MESMA ESCALA DE MEDIDA DOS DADOS
